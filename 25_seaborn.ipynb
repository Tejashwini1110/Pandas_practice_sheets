```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
print(tips.head())
```

```python
import matplotlib.pyplot as plt 
sns.histplot(data=tips,x="total_bill",kde=True,bins=20,color='skyblue')
plt.title("Distribution of bills")
plt.show()
```

```python
sns.kdeplot(data=tips,x='total_bill',shade=True,color='green')
plt.title("KDE plot of Total bills")
plt.show()
```

```python
sns.barplot(data=tips,x='day',y='total_bill',palette='viridis')
plt.title("Average total bill by day") 
plt.show()
```

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex',palette='coolwarm')
plt.title("Box plot of total bills by day and gender")
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',style='time',palette='deep')
plt.title("Scatter plot of total bill vs. Trip")
plt.show()
```

```python
sns.lineplot(data=tips,x='size',y='tip',marker='o',color='green')
plt.title("Line plot of party size vs tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue='sex',palette='husl')
plt.show()
```

```python
numeric_tips=tips.select_dtypes(include=['number'])
correlation_matrix = numeric_tips.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
train=pd.read_csv("train.csv")
train.head()
```

```python
import matplotlib.pyplot as plt
sns.barplot(data=train,x='Pclass',y='Survived',errorbar=None,palette=['red','blue','lavender'])
plt.title('Survival rate by passenger class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.scatterplot(data=train,x='Age',y='Fare',hue='Survived',palette='deep')
plt.title("Relationship between age and fare")
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
selected_data=train[['Age','Fare','Pclass','Survived']]
sns.pairplot(selected_data,hue='Survived',palette='coolwarm',diag_kind='kde')
plt.suptitle('Pairwise Relationships',y=1.02)
plt.show()
```

```python
numeric_train=train.select_dtypes(include=['number'])
correlation_matrix = numeric_train.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f',linewidth=1)
plt.title("correlation Heatmap")
plt.show()
```
